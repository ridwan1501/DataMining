### **1. Pie Chart** 

Melihat presentase dari 5 permainan terbaik

```python
games = ('Dota 2', 'Fortnite', 'Arena of Valor', 'PUBG MOBILE', 'PUBG')
votes = (12, 11, 4, 8, 7)
colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D', '#D53032')
explode = (0.1, 0, 0, 0, 0)

plt.title('Top 5 Games Earnings')
plt.pie(
    votes,
    labels=games,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode,
    shadow=True
    )
plt.show()
```

### **2. Bar Chart**

**Bar Chart** adalah merupakan tools visualisassi yang dapat digunakan untuk membandingkan data kategorikal. Mirip dengan diagram lingkaran, diagram ini dapat digunakan untuk membandingkan kategori data dengan yang lain.

```python
# memuat data
games = ('Dota 2', 'CS:GO', 'PUBG Mobile', 'Rocket League', 'PUBG', 
         'Rainbow Six Siege', 'VALORANT', 'Free Fire', 'Clash of Clans', 
         'Apex Legends','Hearthstone')

all_earnings = (59273109911, 10942810097, 10367055560, 3680407459, 9121454779,
               5683668939, 2182026406, 2534900171, 729812974, 2197631226, 
               4772562831)

# visualisasi bar chart
x_coords = np.arange(len(games))
plt.bar(x_coords, all_earnings)
plt.show()
```

Menunjukkan perbandingan pendapatan dari 11 permainan yang diambil secara acak

```python
# mengurutkan data
games = ('Dota 2', 'CS:GO', 'PUBG Mobile', 'Rocket League', 'PUBG', 
         'Rainbow Six Siege', 'VALORANT', 'Free Fire', 'Clash of Clans', 
         'Apex Legends','Hearthstone')

all_earnings = (59273109911, 10942810097, 10367055560, 3680407459, 9121454779,
               5683668939, 2182026406, 2534900171, 729812974, 2197631226, 
               4772562831)

df = pd.DataFrame({
    'Games': games,
    'All Earnings': all_earnings,
})
df.sort_values(by='All Earnings', inplace=True)

x_coords = np.arange(len(df))
plt.bar(x_coords, df['All Earnings'], tick_label=df['Games'])
plt.xticks(rotation=90)
plt.ylabel('All Earnings (Dollars)')
plt.title('11 Random Pick Games')
plt.show()
```
